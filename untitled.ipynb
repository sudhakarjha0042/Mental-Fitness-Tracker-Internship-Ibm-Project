**IMPORT DATA**

```python
import tensorflow as tf
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import seaborn as sns
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
import sklearn
```

**READING DATA**

```python
from google.colab import drive
drive.mount('/content/drive')

df1=pd.read_csv("/content/drive/MyDrive/Data Sets/mental-and-substance-use-as-share-of-disease.csv")
df2=pd.read_csv("/content/drive/MyDrive/Data Sets/prevalence-by-mental-and-substance-use-disorder.csv")
```

```python
df1.head()
```

```python
df2.head()
```

```python
df=pd.concat(objs=[df2,df1],axis=1)
```

**DATA VISUALIZATION**

```python
corr=df.corr()
plt.figure(figsize=(15,12))
sns.pairplot(df[['Prevalence - Schizophrenia - Sex: Both - Age: Age-standardized (Percent)',
       'Prevalence - Bipolar disorder - Sex: Both - Age: Age-standardized (Percent)',
       'Prevalence - Eating disorders - Sex: Both - Age: Age-standardized (Percent)',
       'Prevalence - Anxiety disorders - Sex: Both - Age: Age-standardized (Percent)',
       'Prevalence - Drug use disorders - Sex: Both - Age: Age-standardized (Percent)',
       'Prevalence - Depressive disorders - Sex: Both - Age: Age-standardized (Percent)',]])

plt.figure(figsize=(15,12))
sns.heatmap(corr)
```

**DATA PREPROCESSING**

```python
df.drop(['Entity','Code','Year'],axis=1,inplace=True)
df=df.fillna(df.mean())

x=df[['Prevalence - Schizophrenia - Sex: Both - Age: Age-standardized (Percent)',
       'Prevalence - Bipolar disorder - Sex: Both - Age: Age-standardized (Percent)',
       'Prevalence - Eating disorders - Sex: Both - Age: Age-standardized (Percent)',
       'Prevalence - Anxiety disorders - Sex: Both - Age: Age-standardized (Percent)',
       'Prevalence - Drug use disorders - Sex: Both - Age: Age-standardized (Percent)',
       'Prevalence - Depressive disorders - Sex: Both - Age: Age-standardized (Percent)',]].to_numpy()

y=df[['DALYs (Disability-Adjusted Life Years) - Mental disorders - Sex: Both - Age: All Ages (Percent)']].to_numpy()

scaler=StandardScaler()
x=scaler.fit_transform(x)

x_train,x_test,y_train,y_test=train_test_split(x,y)
```

**ML IMPLEMENTATION**

```python
ml=RandomForestRegressor()
ml.fit(x_train,y_train)
predicted_values=ml.predict(x_test)
```

**MODEL EVALUATION AND METRICS**

```python
plt.figure(figsize=(15,12))
plt.plot(y_test[:100], label='True Values')
plt.plot(predicted_values[:100], label='Predicted Values')
plt.legend()
plt.title('True vs. Predicted DALYs - Mental Disorders (Test Data)\nMean Square Error: '+str(sklearn.metrics.mean_squared_error(y_test, predicted_values)))
plt.xlabel('Data Point Index')
plt.ylabel('DALYs - Mental disorders - All Ages (%)')
plt.show()
```

**PREDICT USING THE MODEL**

```python
# Assume you have some new data for prediction
new_data = np.array([[0.2, 0.3, 0.4, 0.1, 0.7, 0.6]])  # Replace these values with your new data

# Scale the new data using the same scaler used for training data
new_data_scaled = scaler.transform(new_data)

# Make predictions using the trained model
predicted_new_data = ml.predict(new_data_scaled)

print("Predicted DALYs (Disability-Adjusted Life Years) for the new data:", predicted_new_data)
```

**Scatter plot to visualize the relationship between true and predicted values**

```python
plt.figure(figsize=(10, 8))
plt.scatter(y_test, predicted_values)
plt.xlabel('True DALYs (%)')
plt.ylabel('Predicted DALYs (%)')
plt.title('True vs. Predicted DALYs - Mental Disorders')
plt.show()
```

**Scatter plot to visualize the relationship between all true and predicted values**
